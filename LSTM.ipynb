```python
import pandas as pd

df_denmark=pd.read_csv("denmark_data_new.csv")
```

```python
# Reset the index to remove the 'date' from the index
df_denmark.reset_index(inplace=True)

# Replace '24' in 'hour' column with '00' to handle midnight of the next day
df_denmark['hour'] = df_denmark['hour'].replace(24, 0)

# Create a new 'datetime' column by combining 'date' and 'hour'
df_denmark['datetime'] = pd.to_datetime(df_denmark['date'].astype(str) + ' ' + df_denmark['hour'].astype(str) + ':00')

# Set the new 'datetime' column as the index
df_denmark.set_index('datetime', inplace=True)
```

```python
df_denmark.drop(columns=["index"],inplace=True)
```

```python
df_denmark.head()
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from keras.models import Sequential
from keras.layers import LSTM, Dense
from keras.optimizers import Adam
import matplotlib.pyplot as plt

# Define feature columns and target variable
xgb_features = [
    'hour', 'is_negative_price', 'negative_price_duration', 'is_green_energy',
    'month', 'day_of_week', 'sine_hour', 'cosine_hour', 'is_weekday',
    'price_lag_1', 'price_lag_2', 'price_lag_3', 'price_rolling_7d',
    'price_rolling_30d', 'price_diff_1h', 'price_volatility_24h', 'price_volatility_7d'
]
xgb_target = 'price_eur_per_mwh'

# Separate feature data and target data
X = df_denmark[xgb_features].values
y = df_denmark[xgb_target].values

# Normalize the feature data
feature_scaler = MinMaxScaler(feature_range=(0, 1))
X_scaled = feature_scaler.fit_transform(X)

# Normalize the target data (separately for the target)
target_scaler = MinMaxScaler(feature_range=(0, 1))
y_scaled = target_scaler.fit_transform(y.reshape(-1, 1))  # Reshape to 2D for MinMaxScaler

# Function to create dataset with look-back period
def create_dataset(data, target, look_back=1):
    X_data, y_data = [], []
    for i in range(len(data) - look_back):
        X_data.append(data[i:(i + look_back), :])  # Create sequences of 'look_back' time steps
        y_data.append(target[i + look_back, 0])  # Target variable
    return np.array(X_data), np.array(y_data)

# Prepare the data for sequence prediction
look_back = 10  # Use last 10 time steps to predict the next
X_seq, y_seq = create_dataset(X_scaled, y_scaled, look_back)

# Train-test split (80-20 split)
X_train, X_test = X_seq[:int(0.8 * len(X_seq))], X_seq[int(0.8 * len(X_seq)):]
y_train, y_test = y_seq[:int(0.8 * len(y_seq))], y_seq[int(0.8 * len(y_seq)):]

# Define the LSTM model
def build_lstm_model(input_shape):
    model = Sequential()
    model.add(LSTM(units=64, activation='relu', input_shape=input_shape))
    model.add(Dense(units=1))
    model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error')
    return model

model = build_lstm_model(input_shape=(X_train.shape[1], X_train.shape[2]))

# Train the LSTM model
history = model.fit(X_train, y_train, epochs=100, batch_size=32, validation_data=(X_test, y_test))

# Predict on the test data
y_pred_scaled = model.predict(X_test)

# Inverse scale the predictions and actual values to get the actual price values
y_test_inv = target_scaler.inverse_transform(y_test.reshape(-1, 1))  # Inverse scaling of the actual values (y_test)
y_pred_inv = target_scaler.inverse_transform(y_pred_scaled.reshape(-1, 1))  # Inverse scaling of the predicted values (y_pred)

# Plot the actual vs predicted prices
plt.figure(figsize=(10, 6))
plt.plot(y_test_inv, label='Actual Prices', color='blue')
plt.plot(y_pred_inv, label='Predicted Prices', color='red', linestyle='--')
plt.title('Actual vs Predicted Prices')
plt.xlabel('Time Steps')
plt.ylabel('Price (EUR/MWh)')
plt.legend()
plt.show()

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

def evaluate_model(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_true, y_pred)
    return mae, mse, rmse, r2


# Evaluate the model performance
mae, mse, rmse, r2 = evaluate_model(y_test_inv, y_pred_inv)

# Print evaluation metrics
print(f'Mean Absolute Error (MAE): {mae}')
print(f'Mean Squared Error (MSE): {mse}')
print(f'Root Mean Squared Error (RMSE): {rmse}')
print(f'R2 Score: {r2}')
```

```python
import pandas as pd

# Identify groups of consecutive negative prices
df_denmark["neg_price_group"] = (df_denmark["is_negative_price"].diff(1) != 0).cumsum()

# Filter only negative price events
negative_price_events = df_denmark[df_denmark["is_negative_price"] == 1]
```

```python
# Group by negative price event and count the duration
negative_duration = negative_price_events.groupby("neg_price_group")["hour"].count().reset_index()
negative_duration.rename(columns={"hour": "duration_hours"}, inplace=True)
```

```python
# Merge back to get the end time of negative price events
df_denmark = df_denmark.merge(negative_duration, on="neg_price_group", how="left")

# Identify recovery: Find the first non-negative price after each event
df_denmark["recovery_time"] = df_denmark.groupby("neg_price_group")["price_eur_per_mwh"].transform(
    lambda x: (x >= 0).idxmax() - x.index[0] if (x >= 0).any() else None
)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram of negative price durations
plt.figure(figsize=(8,5))
sns.histplot(negative_duration["duration_hours"], bins=20, kde=True)
plt.title("Distribution of Negative Price Duration")
plt.xlabel("Duration (hours)")
plt.ylabel("Count")
plt.show()

# Scatter plot for Recovery time vs Duration
plt.figure(figsize=(8,5))
sns.scatterplot(data=df_denmark, x="duration_hours", y="recovery_time")
plt.title("Negative Price Duration vs. Recovery Time")
plt.xlabel("Duration of Negative Price (hours)")
plt.ylabel("Recovery Time (hours)")
plt.show()
```
